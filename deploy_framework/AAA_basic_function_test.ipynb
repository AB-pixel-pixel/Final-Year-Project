````python
import json

def extract_last_json(text):
    json_list = []
    length = len(text)
    in_json = False
    start = 0
    stack = 0

    for i in range(length):
        if text[i] == '{':
            if not in_json:
                in_json = True
                start = i
            stack += 1
        elif text[i] == '}':
            stack -= 1
            if stack == 0:
                json_str = text[start:i+1]
                try:
                    json.loads(json_str)
                    json_list.append(json_str)
                except json.JSONDecodeError:
                    # 如果解析失败，跳过这个潜在的json
                    pass
                in_json = False

    if json_list:
        return json_list[-1]
    else:
        return None


# 示例文本
text = """
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
robot 0 first goes to office to explore, then replan to get the environment information. robot 1 first replan to get the environment information, then go grasp target object <bread> (5345043) in the living room.  robot 1 is in the living room, which is near the kitchen, where the items to be transported are. robot 0 is in the office, which is near the bedroom, where the items are to be placed.  This way, robot 0 and robot 1 work together to explore and transport the items. 
robot 0 first goes to office to explore, then replan to get the environment information. robot 1 first replan to get the environment information, then go grasp target object <bread> (5345043) in the living room.  robot 1 is in the living room, which is near the kitchen, where the items to be transported are. robot 0 is in the office, which is near the bedroom, where the items are to be placed.  This way, robot 0 and robot 1 work together to explore and transport the items. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target object <bread> (5345043)"
      ]
    }
  ]
} 
This is the task assignment for the two robots. The first task for robot 0 is to go to the office and replan to get the environment information. The first task for robot 1 is to replan to get the environment information and then go grasp the bread in the living room. This way, the robots can work together to explore and transport the items efficiently. 
In this problem, we have 2 robots. Robot 0 is in the office, and robot 1 is in the living room. The goal is to transport items from the kitchen to the bedroom. The office and living room are the initial rooms that the robots are in. The robots can hold up to 2 items. The environment is unknown, so the robots need to explore and get the environment information first. The items to be transported are 2 orange, 3 apple, 1 banana, 3 bread, and 1 burger. The robots need to go to the kitchen to get the items and then go to the bedroom to place the items. The living room is near the kitchen, so robot 1 can go to the kitchen first. The office is near the bedroom, so robot 0 can go to the bedroom first. This way, the robots can work together to explore and transport the items efficiently. 
The final answer is: 
{
  "reason": "Distribute initial exploration efficiently and transport items to the bed.",
  "robot_id_task_pairs": [
    {
      "robot_id": 0,
      "action_chain": [
        "go to <Office> (3000)",
        "replan"
      ]
    },
    {
      "robot_id": 1,
      "action_chain": [
        "replan",
        "go grasp target

"""

# 提取最后一个json
last_json = extract_last_json(text)
if last_json:
    print("提取的最后一个json为：")
    print(last_json)
else:
    print("未找到有效的json")
````
